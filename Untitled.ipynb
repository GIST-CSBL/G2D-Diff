```python
import pandas as pd
```

```python

nci_data = pd.read_csv("data/drug_response_data/original_nci_drug_response_revised_1129.csv")
train_cell = ['HCT116_LARGE_INTESTINE',
                     'LOXIMVI_SKIN',
                     'K562_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE',
                     'MCF7_BREAST',
                     'CCRFCEM_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE'
                      ]

val_cell = ['EKVX_LUNG', 'SKMEL28_SKIN', 'SKOV3_OVARY', 'NCIH226_LUNG', 'OVCAR4_OVARY']
test_cell = ['TK10_KIDNEY', 'OVCAR5_OVARY', 'HOP92_LUNG', 'SKMEL2_SKIN', 'HS578T_BREAST']


nci_data_train = nci_data[~nci_data['ccle_name'].isin(val_cell + test_cell)]
nci_data_val = nci_data[nci_data['ccle_name'].isin(val_cell)]
nci_data_test = nci_data[nci_data['ccle_name'].isin(test_cell)]



cell2mut = pd.read_csv("data/drug_response_data/original_cell2mut.csv", index_col = 0).rename(columns={'index':'ccle_name'})
cell2cna = pd.read_csv("data/drug_response_data/original_cell2cna.csv", index_col = 0).rename(columns={'index':'ccle_name'})
cell2cnd = pd.read_csv("data/drug_response_data/original_cell2cnd.csv", index_col = 0).rename(columns={'index':'ccle_name'})
drug2smi = pd.read_csv("data/drug_response_data/latent_nci_reconstructed_1129.csv").iloc[:, 0:-1]

```

```python
ev1_dresp = nci_data[nci_data['ccle_name'].isin(train_cell)]
ev2_dresp = nci_data[nci_data['ccle_name'].isin(val_cell)]
ev3_dresp = nci_data[nci_data['ccle_name'].isin(test_cell)]
```

```python
cell2mut_sampled = cell2mut[cell2mut['ccle_name'].isin(train_cell + val_cell + test_cell)]
cell2cna_sampled = cell2cna[cell2cna['ccle_name'].isin(train_cell + val_cell + test_cell)]
cell2cnd_sampled = cell2cnd[cell2cnd['ccle_name'].isin(train_cell + val_cell + test_cell)]

```

```python
cell2mut_sampled.to_csv("cell2mut_sampled.csv", index=False)
cell2cna_sampled.to_csv("cell2cna_sampled.csv", index=False)
cell2cnd_sampled.to_csv("cell2cnd_sampled.csv", index=False)
```

```python
dlist = list(ev1_dresp['drug'].values) + list(ev2_dresp['drug'].values) + list(ev3_dresp['drug'].values)
```

```python
drug2smi_sampled = drug2smi[drug2smi['drug'].isin(set(dlist))]
```
